# Renaming Columns before putting into the Database

```python
df = pd.DataFrame(raw_data)
```

```python
df=raw_data.rename(columns={df.columns[0]: 'area',
                    df.columns[1]: 'date_hour',
                    df.columns[2]: 'biomass',
                    df.columns[3]: 'fossil_brown_coal_lignite',
                    df.columns[4] : 'fossil_coal_derived_gas',
                    df.columns[5]: 'fossil_gas',
                    df.columns[6]:'fossil_hard_coal',
                    df.columns[7]:'fossil_oil',
                    df.columns[8]:'fossil_oil_shale',
                    df.columns[9]:'fossil_peat',
                    df.columns[10]:'geothermal',
                    df.columns[11]:'hydro_pumped_storage',
                    df.columns[12]:'hydro_pumped_storage_consumption',
                    df.columns[13]:'hydro_run_river_poundage',
                    df.columns[14]:'hydro_water_reservoir',
                    df.columns[15]: 'marine',
                    df.columns[16]:'nuclear',
                    df.columns[17]:'other',
                    df.columns[18]:'other_renewable',
                    df.columns[19]:'solar',
                    df.columns[20]:'waste',
                    df.columns[21]:'wind_offshore',
                    df.columns[22]:'wind_onshore'})
```

# Putting the data into the SQLite Database 

```python
import sqlite3
conn = sqlite3.connect('production.db') 
c=conn.cursor()
```

```python
c.execute( 
    """ 
    CREATE TABLE raw_data(
        
                    area TEXT,
                    date_hour TEXT,
                    biomass REAL,
                    fossil_brown_coal_lignite REAL,
                    fossil_coal_derived_gas REAL,
                    fossil_gas REAL,
                    fossil_hard_coal REAL,
                    fossil_oil REAL,
                    fossil_oil_shale REAL,
                    fossil_peat REAL,
                    geothermal REAL,
                    hydro_pumped_storage REAL,
                    hydro_pumped_storage_consumption REAL,
                    hydro_run_river_poundage REAL,
                    hydro_water_reservoir REAL,
                    marine REAL,
                    nuclear REAL,
                    other REAL,
                    other_renewable REAL,
                    solar REAL,
                    waste REAL,
                    wind_offshore REAL,
                    wind_onshore REAL);
"""
)
```

```python
df.to_sql('raw_data', conn, if_exists='append', index=False) #importing the data into the SQL table
```

```python
pd.read_sql("SELECT * FROM raw_data LIMIT 5", conn)
```

```python
conn.close()
```

#### Raw data has been successfully stored in the SQLite database!
